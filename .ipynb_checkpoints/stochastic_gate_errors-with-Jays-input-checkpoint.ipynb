```python
class IonSimError(Exception):
    pass

@dataclass
class AtomicInternalEnergyLevel(ABC):
    """An internal energy level of an atom, i.e., an energy eigenstate of the electronic and nuclear degrees of freedom."""
    n: float
    s: float
    l: float
    j: float

    @staticmethod
    def get_orbital_from_l(l: float):
        """Get the atomic orbital letter from the integer value of its orbital angular momentum."""
        return {0: 'S', 1: 'P', 2: 'D', 3: 'F'}[l]
    
    @property
    @abstractmethod
    def quantum_numbers(self):
        """The quantum numbers of the atomic internal energy level."""
        pass
    
    @property
    @abstractmethod
    def atomic_symbol(self):
        """The atomic symbol of the atomic internal energy level."""
        pass
        
    # @cachedproperty from functools; be careful if internal values change;
    # the cached property won't change with them!

@dataclass
class FineStructureLevel(AtomicInternalEnergyLevel):
    """A fine-structure energy level of an atom."""
    mj: float
    
    @property
    def quantum_numbers(self):
        """The quantum numbers of the fine-structure level."""
        return [self.n, self.s, self.l, self.j, self.mj]
    
    @property
    def atomic_symbol(self):
        """The atmoic symbol of the fine-structure level"""
        # TODO: figure out how to add LaTex rendering to energy-level names
        return f'{AtomicInternalEnergyLevel.get_orbital_from_l(self.l)}{Fraction(self.j)},{Fraction(self.mj)}'

@dataclass
class HyperfineStructureLevel(AtomicInternalEnergyLevel):
    """A hyperfine-structure energy level of an atom."""
    i: float
    f: float
    mf: float
            
    @property
    def quantum_numbers(self):
        """The quantum numbers of the hyperfine-structure level."""
        return [self.n, self.s, self.l, self.j, self.i, self.f, self.mf]
    
    @property
    def atomic_symbol(self):
        """The atmoic symbol of the hyperfine-structure level"""
        return f'{AtomicInternalEnergyLevel.get_orbital_from_l(self.l)}{Fraction(self.j)},{Fraction(self.f)},{Fraction(self.mf)}'
    
class Atom:
    """An atom."""
    @classmethod
    def build(cls, species: str, **kwargs):
        """Build an atom from its atomic level configuration file."""
        with open(f'atomic_level_data/{species}.yaml', 'r') as f:
            data = yaml.safe_load(f)
        assert(data['species'] == species)
        if data['structure'] == 'fine':
            return cls.from_level_data(FineStructureLevel, species, data['quantum_numbers'], **kwargs)
        elif data['structure'] == 'hyperfine':
            return cls.from_level_data(HyperfineStructureLevel, species, data['quantum_numbers'], **kwargs)
        else:
            raise IonSimError(f"Atomic level data for species {species} has {data['structure']} structure. \
                                The structure needs to be fine or hyperfine.")
            
    @classmethod
    def from_level_data(cls, EnergyLevel: type, species: str, quantum_numbers: list[list], atomic_symbols: list[str] = None,
                        qubit_level_atomic_symbols: list[str] = None, **kwargs):
        """Build an atom with fine structure."""     
        levels_list = [EnergyLevel(*qns) for qns in quantum_numbers]
        if atomic_symbols is None:
            levels_dict = {level.atomic_symbol: level for level in levels_list}
        else:
            levels_dict = {level.atomic_symbol: level for level in levels_list if level.atomic_symbol in atomic_symbols}
            
        if qubit_level_atomic_symbols is not None:
            if not all([symbol in levels_dict for symbol in qubit_level_atomic_symbols]):
                raise IonSimError(f'The qubit levels {qubit_level_atomic_symbols} are not included in the full set '
                                  f'of energy levels {list(levels_dict)}.') 
                
        if qubit_level_atomic_symbols is None:
            qubit_levels_dict = None
        else:
            qubit_levels_dict = {level.atomic_symbol: level for level in levels_list if level.atomic_symbol in qubit_level_atomic_symbols}
        return cls(species, levels_dict, qubit_levels_dict)
        
    def __init__(self, species: str, internal_energy_levels: dict, qubit_levels: dict = None):
        # TODO: auto-generate atomic properties, like mass, from the species
        if qubit_levels is not None:
            if not all([symbol in list(internal_energy_levels) for symbol in list(qubit_levels)]):
                raise IonSimError(f'The qubit levels {list(qubit_levels)} are not included in the full set '
                                  f'of energy levels {list(internal_energy_levels)}.') 
        self.species = species
        self.internal_energy_levels = internal_energy_levels
        self.qubit_levels = qubit_levels
    
    @property
    def zero(self):
        """The energy level of the atom that represents qubit level zero."""
        if self.qubit_levels is None:
            return
        else:
            return list(self.qubit_levels.values())[0]
    
    @property
    def one(self):
        """The energy level of the atom that represents qubit level one."""
        if self.qubit_levels is None:
            return
        else:
            return list(self.qubit_levels.values())[1]
         
        
class InternalEnergyLevel:
    """An internal energy level of a collection of atoms."""
    def __init__(self, atomic_levels: list[AtomicInternalEnergyLevel]):
        self.atomic_levels = atomic_levels
        self.name = ' '.join([atomic_level.name for atomic_level in atomic_levels])
        self.atomic_symbol = ' '.join([atomic_level.atomic_symbol for atomic_level in atomic_levels])
        
```

```python
class SpinBasis: # NB: we don't need the parentheses. e.g. SpinBasis()
    """A basis of spin states (i.e., internal states) for a collection of atomic qubits."""
    def __init__(self, atoms: list[Atom]): 
        self.atoms = atoms
        self.states = None
        self.state_names = None
        self.vectors = None
            
    def enlarge_matrix(self, matrix, current_atoms: list[Atom]): # TODO: figure out typing for a matrix
        """Enlarge the dimension of a matrix in a basis of some atoms to include all atoms in the basis."""
        if len(current_atoms) == 1:
            return self.enlarge_one_atom_matrix(matrix, *current_atoms)
        else:
            raise ValueError('Enlarging a matirx for more than one atom is not currently implemented.')
            
    def enlarge_one_atom_matrix(self, matrix, current_atom: Atom):
        """Enlarge the dimension of a matrix in a basis of one atom to include all atoms in the basis."""
        large_matrix = np.array([1])
        for atom in self.atoms:
            if atom == current_atom:
                large_matrix = np.kron(large_matrix, matrix)
            else:
                large_matrix = np.kron(large_matrix, np.eye(len(matrix)))
        return large_matrix
    
    def enlarge_matrix_function(self, matrix_function, current_atoms: list[Atom]):
        """Enlarge the dimension of a matrix function in a basis of some atoms to include all atoms in the basis."""
        # TODO: current wrapping does not preserve inspect.getfullargspec!
        @wraps(matrix_function)
        def wrapper(*args, **kwargs):
            if len(current_atoms) == 1:
                return self.enlarge_one_atom_matrix(matrix_function(*args, **kwargs), *current_atoms)
            else:
                raise ValueError('Enlarging a matirx for more than one atom is not currently implemented.')
        return wrapper
        
class StandardSpinBasis(SpinBasis):
    """A basis in which each basis state corresponds to an internal energy level of a collection of atomic qubits."""
    def __init__(self, *args, **kwargs):  
        SpinBasis.__init__(self, *args, **kwargs)
        atomic_levels_list = list(it.product(*[atom.qubit_levels for atom in self.atoms]))
        self.states = [InternalEnergyLevel(atomic_levels) for atomic_levels in atomic_levels_list]
        self.vectors = list(np.eye(len(self.states)))
        self.state_names = []
        self.state_atomic_symbols = []
        for state, vector in zip(self.states, self.vectors):
            state.vector = vector
            self.state_names += [state.name]
            self.state_atomic_symbols += [state.atomic_symbol]
        # self.state_names = [state.name for state in self.states]
        # self.state_atomic_symbols = [state.atomic_symbol for state in self.states]
        # TODO: Is it actually faster to create these lists in the same loop, as done above? Or are they compiled to the
        # same code at runtime? If they're the same speed, the list comprehension method is less lines of code - so prehaps
        # more readable. In this example, the runtime is trivial, but I should figure this out for future reference.
        # NB: LIST COMPREHENSIONS ARE FASTER! EVEN MULTIPLE ONES IN ROW.
        # NB: USE list.append(). It's faster! Also use .extend() for adding multiple entries instead of "+"
        
    def compute_superoperator_from_unitary_operator(self, unitary_operator):
        """Compute a superoperator from a unitary operator the column-stacked basis"""
        #TODO: make this function reflect the choice of basis
        return np.kron(unitary_operator.conj(), unitary_operator)
    
    def create_superoperator_function_from_unitary_operator_function(self, unitary_operator_function):
        """Create a superoperator function from a unitary operator function in the column-stacked basis."""
        # TODO: current wrapping does not preserve inspect.getfullargspec!
        @wraps(unitary_operator_function)
        def wrapper(*args, **kwargs):
            return self.compute_superoperator_from_unitary_operator(unitary_operator_function(*args, **kwargs))
        return wrapper
    
```

```python
class SpinMatrix():
    """A quantm state or operator in a basis of spin states."""
    def __init__(self, basis: SpinBasis):
        self.basis = basis
        
class SpinState(SpinMatrix):
    """A quantum state in a basis of spin states."""
    
    @classmethod
    def from_coefs(cls, coefs: list[float])
        return cls() # call class constructor in useful way
    
    #NB: Or create a new instance of things to keep immutability
    
    def __init__(self, *args, **kwargs):
        SpinMatrix.__init__(self, *args, **kwargs)
        self.name = None
        self.wavefunction = None
        self.density_matrix = None
        self.supervector = None
        
    # NB: Jay isn't bothered by cryptic name, in this case
    def from_coefs(self, coefs: list[float]): # TODO: figure out typing for arrays
        """Build the state from a list of basis-state coefficients."""
        # TODO: is starting the name of a method like this one with the word "from" sensible?
        # I guess we could start the name with "build" instead. I like the compactness of
        # starting the name with "from, but perhaps the name is too cryptic.
        # Speaking of being cryptic, is "coefs" too cryptic? Is "from_basis_state_coefficients" better?
        self.wavefunction = self.compute_wavefunction_from_coefs(coefs)
        self.density_matrix = self.compute_density_matrix_from_wavefunction(self.wavefunction)
        self.supervector = self.compute_supervector_from_density_matrix(self.density_matrix)
        return self
    
    def from_wavefunction(self, wavefunction): # TODO: figure out typing for arrays
        """Build the state from a wavefunction."""
        self.wavefunction = wavefunction
        self.density_matrix = self.compute_density_matrix_from_wavefunction(wavefunction)
        self.supervector = self.compute_supervector_from_density_matrix(self.density_matrix)
        return self
    
    def from_density_matrix(self, density_matrix): # TODO: figure out typing for arrays
        """Build the state from a density matrix."""
        self.wavefunction = None
        self.density_matrix = density_matrix
        self.supervector = self.compute_supervector_from_density_matrix(density_matrix)
        return self
    
    def from_supervector(self, supervector): # TODO: figure out typing for arrays
        """Build the state from a supervector."""
        self.wavefunction = None
        self.density_matrix = self.compute_density_matrix_from_supervector(supervector)
        self.supervector = supervector
        return self
        
    def compute_wavefunction_from_coefs(self, coefs: list[float]):
        """Compute a wavefunction from its basis coefficients."""
        # TODO: Functions like this shouldn't be called by an instance of the class.
        # Should we prepend the name with "_" or do something else?
        assert(len(coefs) == len(self.basis.states))
        coefs = list(np.array(coefs)/np.linalg.norm(coefs))
        return sum([state.vector*coef for state,coef in zip(self.basis.states, coefs)])
    
    def compute_density_matrix_from_wavefunction(self, wavefunction): # TODO: figure out typing for arrays
        """Compute a density matrix from a wavefunction, i.e., a pure state."""
        assert(len(wavefunction) == len(self.basis.states))
        assert(abs((wavefunction.conj().T).dot(wavefunction)-1) < 1e-12)
        return np.outer(wavefunction, wavefunction.conj().T)
    
    def compute_supervector_from_density_matrix(self, density_matrix): # TODO: generalize for any basis
        """Compute a column-stacked supervector from a density matrix."""
        assert(len(density_matrix) == len(self.basis.states))
        assert(density_matrix.shape == (len(self.basis.states), len(self.basis.states)))
        return (density_matrix.T).flatten()
    
    def compute_density_matrix_from_supervector(self, supervector): # TODO: generalize for any basis
        """Compute a process matrix from a column-stacked supervector."""
        dimension = len(self.basis.states)
        assert(len(supervector) == dimension**2)
        return (supervector.reshape(dimension, dimension)).T
    
    def propagate_using_process_matrix(self, process_matrix):
        """Propagate the state by applying a process matrix to the supervector."""
        self.wavefunction = None
        self.supervector = process_matrix.dot(self.supervector)
        self.density_matrix = self.compute_density_matrix_from_supervector(self.supervector)
    
    def compute_state_fidelity(self, target_density_matrix):
        """Compute the state fidelity with respect to a target density matrix."""
        return np.trace(np.dot(self.density_matrix, target_density_matrix)).real
    
    def compute_basis_state_probabilities(self):
        """Compute the probability of measuring each basis state."""
        probabilities = []
        for vector in self.basis.vectors:
            projector = np.outer(vector, vector.conj().T)
            probabilities += [np.trace(projector.dot(self.density_matrix)).real]
        return probabilities
```

```python
class Noise():
    """A stochastic fluctuation with a known probability density function."""
    def __init__(self, probability_density_function: callable, domain_arguments: list):
        self.probability_density_function = probability_density_function
        self.domain_arguments = domain_arguments

    def add_noise_to_function(self, function: callable, parameter_index: int):
        """Replace a function with one averaged over the noise probability density."""
        @wraps(function)
        def wrapper(*args, **kwargs):
            function_arguments = np.array([[*args]]*len(self.domain_arguments))
            function_arguments[:,parameter_index] += self.domain_arguments
            function_values = [function(*arguments) for arguments in function_arguments]
            probs = [self.probability_density_function(darg) for darg in self.domain_arguments]
            ys = np.array([p*fv for p, fv in zip(probs, function_values)])
            return trapz_for_matrix(ys, self.domain_arguments)
        return wrapper
```

```python
def trapz_for_matrix(ys, xs, *args, **kwargs): # TODO: figure out the correct type annotation for an array.
    """Apply scipy.integrate.trapz to a matrix of integrands."""
    # TODO: Should functions like this be part of a "math" module?
    num_rows, num_columns = ys[0].shape
    integral = np.zeros((num_rows, num_columns), dtype='complex')
    for row in range(num_rows):
        for column in range(num_columns):
            integrands = np.array([y[row, column] for y in ys])
            integral[row, column] = sp.integrate.trapz(integrands, xs, *args, **kwargs)
    return integral

def gaussian(x: float, sigma: float):
    """A Gaussian function."""
    # TODO: the questions in the trapz_for_matrix function apply here also
    norm = 1/np.sqrt(2*np.pi*sigma**2)
    return norm * np.exp(-x**2/(2*sigma**2))

def exponential_decay(t: float, tau: float):
    """A decaying exponential function."""
    norm = 1/(2*tau)
    return norm * np.exp(-abs(t)/tau)

def constant_with_cutoff(x: float, x0: float):
    norm = 1/(2*x0)
    if abs(x) <= x0:
        return norm
    else:
        return 0.0

probability_density_functions = {'Gaussian': gaussian,
                                 'Exponential decay': exponential_decay,
                                 'Constant with cutoff': constant_with_cutoff}
# The purpose of this dictionary is for users to be able to call up this kind of function by name.
# What should we use instead of a dictionary?
# Should we use a class like the one below?

# class GaussianNoise(Noise):
#     """Noise with a Gaussian probability function."""
#     def __init__(self *args, **kwargs):
#         def gaussian_function(x, sigma):
#             norm = 1/np.sqrt(2*np.pi*sigma**2)
#             return norm*np.exp(-x**2/(2*sigma**2))
#         Noise.__init__(self, gaussian_function, *args, **kwargs)
```

```python
# TODO: the questions below the probability_density_functions dictionary apply here also.
# Where should we store all of the information below??

pauli_spin_matrices = {
    'I': np.array([[1, 0], [0, 1]]),
    'X': np.array([[0, 1], [1, 0]]),
    'Y': np.array([[0, -1j], [1j, 0]]),
    'Z': np.array([[1, 0], [0, -1]])}

def rotation(phi: float, theta: float):
    """A single-qubit rotation on the Bloch sphere."""
    phase_factor = np.exp(1j*theta/2)
    sig_phi = np.cos(phi)*pauli_spin_matrices['X'] + np.sin(phi)*pauli_spin_matrices['Y']
    return phase_factor * (np.cos(theta/2)*pauli_spin_matrices['I'] - 1j*np.sin(theta/2)*sig_phi)

def idle(duration: float):
    """A single-qubit idle gate."""
    return pauli_spin_matrices['I']

def phase(detuning, duration):
    phi = detuning*duration
    return np.array([[1, 0],
                     [0, np.exp(1j*phi)]])

unitary_gate_functions = {'R': rotation,
                          'Idle': idle,
                          'Phase': phase}

unitary_gates = {'Px': unitary_gate_functions['R'](0, np.pi),
                 'Sx': unitary_gate_functions['R'](0, np.pi/2)}
 
```

```python
class SpinProcess(SpinMatrix):
    """A quantum process represented in a basis of spin states."""
    
    some_dict = {'key': stuff}
    value = 1 # never want to change; and all instances share
    
    @staticmethod
    def rotation():
    
    @classmethod
    def rotation(cls):
    
    # NB: We made this class because some attributes and methods are the same for a gate and a circuit.
    def __init__(self, *args, **kwargs):
        SpinMatrix.__init__(self, *args, **kwargs)
        self.name = None
        self.unitary_matrix = None
        self.process_matrix = None
        
    def compute_average_process_fidelity(self, target_process_matrix):
        """Compute the average process fidelity with respect to a target process matrix."""
        total = 0
        for basis_state_vector in np.eye(len(self.process_matrix), dtype='complex'):
            final_state_vector = self.process_matrix.dot(basis_state_vector)
            target_state_vector = target_process_matrix.dot(basis_state_vector)
            total += np.dot(target_state_vector.conj().T, final_state_vector).real
        return total/len(self.process_matrix)
    
class SpinGate(SpinProcess):
    """A quantum gate represented in a basis of spin states."""
    def __init__(self, target_atoms: list[Atom], *args, **kwargs):
        SpinProcess.__init__(self, *args, **kwargs)
        self.target_atoms = target_atoms
        
        self.process_matrix_function = None
        self.parameter_names = None
        self.arguments = None
        
        self.noise = None
        self.noisy_parameter_name = None
        self.noisy_parameter_index = None
    
    def from_name_of_unitary_gate(self, name: str):
        """Build the gate from the name of a unitary gate."""
        self.name = name
        self.unitary_matrix = self.basis.enlarge_matrix(unitary_gates[name], self.target_atoms)
        self.process_matrix = self.basis.compute_superoperator_from_unitary_operator(self.unitary_matrix)
        return self
    
    def from_name_and_arguments_of_unitary_gate_function(self, name: str, arguments: list):
        """Build the gate from the name and arguments of a unitary gate function."""
        # TODO: is this method name too long/wordy?
        return self.from_unitary_gate_function_and_arguments(unitary_gate_functions[name], arguments, name = name)
    
    def from_unitary_gate_function_and_arguments(self, unitary_gate_function: callable, arguments: list, name: str = None):
        """Build the gate from a custom unitary gate function and its arguments."""
        self.name = name
        self.arguments = arguments
        self.parameter_names = inspect.getfullargspec(unitary_gate_function)[0]
        # TODO: the wrapping below does not preserve inspect.getfullargspec! Can we fix it?
        function = self.basis.enlarge_matrix_function(unitary_gate_function, self.target_atoms)
        self.unitary_matrix = function(*arguments)
        self.process_matrix_function = self.basis.create_superoperator_function_from_unitary_operator_function(function)
        self.process_matrix = self.process_matrix_function(*arguments)
        return self
    
    def add_noise(self, noise: Noise, parameter_name: str):
        """Rebuild the gate and add noise to one parameter. This gate-level noise parameter is
        contant during a single gate and fluctuates gate to gate."""
        # TODO: is this method name sensible? I guess we could start the name with "rebuild" instead.
        # NB: the order of "build" and "rebuild" methods does matter. A "build" method needs to be done first.
        self.noise = noise
        self.noisy_parameter_name = parameter_name # TODO: allow for noise on more than one parameter
        assert(self.process_matrix_function is not None)
        assert(self.parameter_names is not None)
        assert(self.arguments is not None)
        assert(parameter_name in self.parameter_names)
        self.noisy_parameter_index = self.parameter_names.index(parameter_name)
        self.process_matrix_function = self.noise.add_noise_to_function(self.process_matrix_function, self.noisy_parameter_index)
        self.process_matrix = self.process_matrix_function(*self.arguments)
        return self
    
class SpinCircuit(SpinProcess):
    """A quantum circuit (i.e., a series of gates) in a basis of spin states."""
    def __init__(self, gates: list[SpinGate], *args, **kwargs): # TODO: check if everything would work for non-unitary gates
        SpinProcess.__init__(self, *args, **kwargs)
        self.gates = gates
        self.process_matrix = self.combine_process_matrices([gate.process_matrix for gate in gates])
        
# TODO: figure out if a circuit process matrix function would be useful or not.
#         if all([gate.process_matrix_function is None for gate in gates]):
#             self.process_matrix_function = None
#         else:
#             self.process_matrix_function = self.combine_process_matrix_functions([gate.process_matrix_function for gate in gates])
            
        self.noise = None
        self.noisy_parameter_name = None
        self.noisy_parameter_index_for_each_gate = None
        
    def combine_process_matrices(self, process_matrices: list[SpinMatrix]):
        """Combine a series of process matrices (in chronological order) into a single process matrix for the whole circuit."""
        if len(process_matrices) == 1:
            return process_matrices[0]
        else:
            return np.linalg.multi_dot(process_matrices[::-1])
        
#     def combine_process_matrix_functions(self, process_matrix_functions: list[callable]):
#         return None
        
    def add_noise(self, noise: Noise, parameter_name: str):
        """Rebuild the circuit and add noise to one parameter. This circuit-level noise parameter
        is constant during a single circuit and fluctuates circuit to circuit."""
        self.noise = noise
        self.noisy_parameter_name = parameter_name
        self.noisy_parameter_index_for_each_gate = []
        for gate in self.gates:
            if gate.parameter_names is not None and (parameter_name in gate.parameter_names):
                self.noisy_parameter_index_for_each_gate += [gate.parameter_names.index(parameter_name)]
            else:
                self.noisy_parameter_index_for_each_gate += [None]
        assert(any([npi is not None for npi in self.noisy_parameter_index_for_each_gate]))
        process_mats = []
        for darg in self.noise.domain_arguments:
            pmat_list = []
            for ig, gate in enumerate(self.gates):
                parameter_index = self.noisy_parameter_index_for_each_gate[ig]
                if parameter_index is not None:
                    nominal_gate_args = list(gate.arguments)
                    gate.arguments[parameter_index] += darg
                    pmat = gate.process_matrix_function(*gate.arguments)
                    gate.arguments = list(nominal_gate_args)
                else: 
                    pmat = gate.process_matrix
                pmat_list += [pmat]
            process_mats += [self.combine_process_matrices(pmat_list)]
        probs = [self.noise.probability_density_function(darg) for darg in self.noise.domain_arguments]
        ys = np.array([p*chi for p, chi in zip(probs, process_mats)])
        self.process_matrix = trapz_for_matrix(ys, self.noise.domain_arguments) 
        return self
    
```

```python
sigma = 500 * (2*np.pi)
def pdf(x):
    sig = sigma
    f = probability_density_functions['Gaussian']
    return f(x, sig)
dargs = np.linspace(-10*sigma, 10*sigma, 201)
# pds = np.array([pdf(darg) for darg in dargs])
# plt.plot(dargs, pds)
# plt.show()
# print(sp.integrate.trapz(pds, dargs))

noise = Noise(pdf, dargs)

def idle_then_rotation(detuning, idle_duration, phi, theta):
    phi += detuning*idle_duration
    return rotation(phi, theta)

idle_durations = np.linspace(0, 2*np.pi/sigma, 41)
contrasts = []
for idle_duration in idle_durations:
    
    phis = np.linspace(0, 2*np.pi, 41)
    fidelities = []
    for phi in phis:
        ramsey = SpinCircuit([
            SpinGate([atom_a], basis).from_name_of_unitary_gate('Sx'),
            SpinGate([atom_a], basis).from_unitary_gate_function_and_arguments(
                idle_then_rotation, [0, idle_duration, phi, np.pi/2]).add_noise(noise, 'detuning')],
            basis)
        fidelities += [ramsey.compute_average_process_fidelity(Px.process_matrix)]
#     plt.plot(phis, fidelities)
#     plt.ylabel('Fidelity')
#     plt.xlabel('Phase (rad.)')
#     plt.show()
#     i0 = list(phis).index(0)
#     ipi = list(phis).index(np.pi)
#     i2pi = list(phis).index(2*np.pi)
#     print(fidelities[i0], fidelities[ipi], fidelities[i2pi])
    contrasts += [abs(max(fidelities) - min(fidelities))]
    
plt.plot(np.array(idle_durations)*1e3, contrasts)
plt.ylabel('Ramsey Fringe Contrast')
plt.xlabel(r'Ramsey Delay (ms)')
plt.show()
```

```python
def f(t, s):
    return np.exp(-t**2*s**2/2)
params, covars = sp.optimize.curve_fit(f, idle_durations, contrasts, p0=[sigma])
print(params[0]/(2*np.pi), covars)
plt.plot(np.array(idle_durations)*1e3, contrasts, '.')
plt.plot(np.array(idle_durations)*1e3, f(idle_durations, *params), '-')
plt.ylabel('Ramsey Fringe Contrast')
plt.xlabel(r'Ramsey Delay (ms)')
plt.show()
print(sigma/(2*np.pi))
```
